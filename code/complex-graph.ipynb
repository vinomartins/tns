Considere $f(z) = (z-a)^{n_1}(z-b)^{n_2}$, $n_1,n_2 \geq 1$, $a\neq b \in \mathbb{C}$.
O código a seguir calcula as raízes de $\frac{df}{dz}$, a saber, $\{a,b,\zeta = \frac{n_2a+n_1b}{n_1+n_2}\}$. A célula a seguir plota o gráfico de $f(z)$ usando o esquema de cores e os segmentos ligando os pontos $a$ e $\zeta$, e $\zeta$ e $b$.

**Teorema**: Seja $l_{\rm parcial}$ a distância $a$ a $\zeta$ e $l_{\rm total}$ a distância de $a$ a $b$,
então
$$
\frac{l_{\rm parcial}}{l_{\rm total}} = \frac{n_2}{n_1+n_2}
$$
Em particular, se $n_1 = n_2$, a raiz $\zeta$ está no ponto médio entre $a$ e $b$.


***obs.:*** Como não é possível prever a ordem que as raízes aparecem no método `roots()`, os segmentos podem não coincidirem com o previsto.
